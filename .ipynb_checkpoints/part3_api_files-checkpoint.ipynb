# Task 1 — File Read & Write Basics 

## Part A - Write

```python
# write mode to create and write the file
with open("python_notes.txt", "w", encoding="utf-8") as f:
    f.write("Topic 1: Variables store data. Python is dynamically typed.\n")
    f.write("Topic 2: Lists are ordered and mutable.\n")
    f.write("Topic 3: Dictionaries store key-value pairs.\n")
    f.write("Topic 4: Loops automate repetitive tasks.\n")
    f.write("Topic 5: Exception handling prevents crashes.\n")

print("File written successfully.")
```

```python
# append mode adds to the file without deleting what's already there
with open("python_notes.txt", "a", encoding="utf-8") as f:
    f.write("Topic 6: Functions make code reusable.\n")
    f.write("Topic 7: You can import modules to add more features.\n")

print("Lines appended.")
```

## Part B - Read

```python
# opening file in read mode
with open("python_notes.txt", "r", encoding="utf-8") as f:
    lines = f.readlines()

# printing each line numbered and stripping the \n at the end
for i in range(len(lines)):
    print(str(i + 1) + ".", lines[i].strip())

print()
```

```python
# total number of lines
print("Total lines:", len(lines))
```

```python
# asking user for keyword and searching
keyword = input("Enter a keyword to search: ")

found = False
for line in lines:
    if keyword.lower() in line.lower():
        print(line.strip())
        found = True

if found == False:
    print("No lines found with that keyword.")
```

# Task 2 — API Integration

## Step 1 - Fetch and Display Products

```python
import requests
import pandas as pd
```

```python
# Step 1 - Fetch and Display
response = requests.get("https://dummyjson.com/products?limit=20")
data = response.json()
products = data["products"]
```

```python
# converting to dataframe - instant table
df = pd.DataFrame(products)[["id", "title", "category", "price", "rating"]]

# formatted table
from tabulate import tabulate
print(tabulate(df, headers="keys", tablefmt="pipe", showindex=False))
```

```python


filtered_table = df[df['rating']>=4.5]

filtered_sorted_table = filtered_table.sort_values(by='price',ascending=False)
```

```python
print(tabulate(filtered_sorted_table, headers="keys", tablefmt="pipe", showindex=False))
```

## Task 3 — Exception Handling

## Part A — Guarded Calculator

```python
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Error: Cannot divide by zero"
    except TypeError:
        return "Error: Invalid input types"

print(safe_divide(10, 2))
print(safe_divide(10, 0))
print(safe_divide("ten", 2))
```

# Part B — Guarded File Reader

```python
def read_file_safe(filename):
    try:
        with open(filename, "r", encoding="utf-8") as f:
            content = f.read()
        return content
    except FileNotFoundError:
        print("Error: File '" + filename + "' not found.")
    finally:
        print("File operation attempt complete.")

print(read_file_safe("python_notes.txt"))
print(read_file_safe("ghost_file.txt"))
```

## Part C — Robust API Calls

```python
# Step 1 - Fetch and Display

try:
    response = requests.get("https://dummyjson.com/products?limit=20", timeout=5)
    data = response.json()
    products = data["products"]

    df = pd.DataFrame(products)[["id", "title", "category", "price", "rating"]]
    print(tabulate(df, headers="keys", tablefmt="pipe", showindex=False))

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(e)

print()

#Step 2 - Filter and Sort

try:
    df_filtered = df[df["rating"] >= 4.5]
    df_sorted = df_filtered.sort_values(by="price", ascending=False)
    print("Filtered (rating >= 4.5) sorted by price:")
    print(tabulate(df_sorted, headers="keys", tablefmt="pipe", showindex=False))

except Exception as e:
    print(e)

print()

#Step 3 - Laptops

try:
    laptop_response = requests.get("https://dummyjson.com/products/category/laptops", timeout=5)
    laptop_data = laptop_response.json()
    laptops = laptop_data["products"]

    df_laptops = pd.DataFrame(laptops)[["title", "price"]]
    print("Laptops:")
    print(tabulate(df_laptops, headers="keys", tablefmt="pipe", showindex=False))

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(e)

print()

#Step 4 - POST

try:
    new_product = {
        "title": "My Custom Product",
        "price": 999,
        "category": "electronics",
        "description": "A product I created via API"
    }

    post_response = requests.post("https://dummyjson.com/products/add", json=new_product, timeout=5)
    print("POST response:")
    print(post_response.json())

except requests.exceptions.ConnectionError:
    print("Connection failed. Please check your internet.")
except requests.exceptions.Timeout:
    print("Request timed out. Try again later.")
except Exception as e:
    print(e)
```

## Part D — Input Validation Loop

```python
### while True:
    user_input = input("Enter a product ID to look up (1-100), or 'quit' to exit: ")

    if user_input == "quit":
        break

    if user_input.isnumeric() == False:
        print("Warning: enter a valid number.")
        continue

    if int(user_input) < 1 or int(user_input) > 100:
        print("Warning: ID must be between 1 and 100.")
        continue

    response = requests.get("https://dummyjson.com/products/" + user_input, timeout=5)

    if response.status_code == 404:
        print("Product not found.")
    elif response.status_code == 200:
        p = response.json()
        print("Title:", p["title"])
        print("Price: $" + str(p["price"]))

    print()
```

# Task 4 — Logging to File

```python
import requests
from datetime import datetime

# function to write to error log
def log_error(location, error_type, message):
    timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    entry = "[" + timestamp + "] ERROR in " + location + ": " + error_type + " — " + message + "\n"
    with open("error_log.txt", "a", encoding="utf-8") as f:
        f.write(entry)

# trigger 1 - ConnectionError using unreachable URL
try:
    requests.get("https://this-host-does-not-exist-xyz.com/api", timeout=5)
except requests.exceptions.ConnectionError as e:
    log_error("fetch_products", "ConnectionError", str(e))
    print("ConnectionError logged.")
except requests.exceptions.Timeout:
    log_error("fetch_products", "Timeout", "Request timed out")
    print("Timeout logged.")

# trigger 2 - HTTP 404 by requesting product ID 999
response = requests.get("https://dummyjson.com/products/999", timeout=5)
if response.status_code != 200:
    log_error("lookup_product", "HTTPError", str(response.status_code) + " Not Found for product ID 999")
    print("HTTPError logged.")

# reading and printing the full log
print()
print("--- error_log.txt ---")
with open("error_log.txt", "r", encoding="utf-8") as f:
    print(f.read())
```
